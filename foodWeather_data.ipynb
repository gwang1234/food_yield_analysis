```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
# 강수량 
water = pd.read_excel('강수량.xlsx', index_col=0)

# 헤더를 인덱스로 변경 + 컬럼명 지정 
water = water.stack(level=0, future_stack=True).reset_index()
water.columns = ['행정구역별', '년도', '강수량']

# 인덱스 설정 
water = water.set_index(['년도', '행정구역별'])

# 정렬 
water = water.sort_values(by='년도', ascending=False)

# 인덱스 값 수정 (두 번째 레벨만)
water.index = water.index.map(lambda x: (x[0], x[1].replace('전북특별자치도', '전라북도').replace('제주특별자치도', '제주도')
                                         .replace('강원특별자치도', '강원도').replace('전국(평균)', '계')))
water.head(20)
```

```python
# 기온
temperature = pd.read_excel('기온.xlsx', index_col=0)

# 헤더를 인덱스로 변경 + 컬럼명 지정 
temperature = temperature.stack(level=0, future_stack=True).reset_index()
temperature.columns = ['행정구역별', '년도', '평균기온']

# 인덱스 설정 
temperature = temperature.set_index(['년도', '행정구역별'])

# 정렬 
temperature = temperature.sort_values(by='년도', ascending=False)

# 인덱스 값 수정 (두 번째 레벨만)
temperature.index = temperature.index.map(lambda x: (x[0], x[1].replace('전북특별자치도', '전라북도').replace('제주특별자치도', '제주도')
                                         .replace('강원특별자치도', '강원도').replace('전국(평균)', '계')))
temperature.head(20)
```

```python
# 지하수 중금속농도
quality = pd.read_excel('지역_수질.xlsx', header=[0,1], index_col=0)

# 특정 행, 열 삭제 
quality = quality.drop(columns=[('구분(2)', '구분(2)')])
quality = quality[quality.index.notna()]

# 헤더를 인덱스로 변경 + 컬럼명 지정 
quality = quality.stack(level=0, future_stack=True).reset_index()
quality.columns = ['행정구역별', '년도', 'pH', '총대장균군 (군수/100mL)', '질산성질소NO3-N (mg/L)', '염소이온Cl- (mg/L)']

# 인덱스 설정 
quality = quality.set_index(['년도', '행정구역별'])

# 정렬 
quality = quality.sort_values(by='년도', ascending=False)

# 인덱스 값 수정 (두 번째 레벨만)
quality.index = quality.index.map(lambda x: (x[0], x[1].replace('전북특별자치도', '전라북도').replace('제주특별자치도', '제주도')
                                         .replace('강원특별자치도', '강원도').replace('전국(평균)', '계')))
quality
```

```python
# 지역별 연 강수량
plt.figure(figsize=(14, 6))
plt.title("지역별 연 강수량 - 강원도")

region = water.xs('강원도', level='행정구역별').reset_index()
plt.bar(region['년도'], region['강수량'], width=0.6, color = '#1a9def')
plt.ylim(800, 2400)
plt.ylabel("강수량")
plt.legend()
plt.show()
```

```python
# 지역별 연 평균기온 
# TODO: 이게 과연 필요할까 그냥 납둘까 
plt.figure(figsize=(14, 6))
plt.title("지역별 연 평균기온 - 강원도")

region = temperature.xs('강원도', level='행정구역별').reset_index()
plt.plot(region['년도'], region['평균기온'], color = "#ef611a", marker='o')
plt.ylabel("평균기온")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

region_data = quality.xs('강원도', level='행정구역별').reset_index()

cols_to_numeric = ['pH', '총대장균군 (군수/100mL)', '질산성질소NO3-N (mg/L)', '염소이온Cl- (mg/L)']
for col in cols_to_numeric:
    region_data[col] = pd.to_numeric(region_data[col], errors='coerce')

fig, axs = plt.subplots(2, 2, figsize=(14, 10), sharex=True)
plt.suptitle("지역별 지하수 중금속 농도 - 강원도", fontsize = 16)

# 1) pH
axs[0, 0].plot(region_data['년도'], region_data['pH'], marker='o', color='blue')
axs[0, 0].set_title('pH')
axs[0, 0].set_ylabel('pH')

# 2) 총대장균군
axs[0, 1].bar(region_data['년도'], region_data['총대장균군 (군수/100mL)'], color='orange', alpha=0.7)
axs[0, 1].set_title('총대장균군')
axs[0, 1].set_ylabel('총대장균군 (군수/100mL)')

# 3) 염소이온
axs[1, 0].bar(region_data['년도'], region_data['염소이온Cl- (mg/L)'], color='green')
axs[1, 0].set_title('염소이온')
axs[1, 0].set_ylim(10, 180)
axs[1, 0].set_ylabel('염소이온Cl- (mg/L)')

# 4) 질산성질소 NO3-N
axs[1, 1].plot(region_data['년도'], region_data['질산성질소NO3-N (mg/L)'], marker='o', color='red')
axs[1, 1].set_title('질산성질소 NO3-N')
axs[1, 1].set_ylabel('질산성질소NO3-N (mg/L)')


plt.tight_layout()
plt.show()
```
